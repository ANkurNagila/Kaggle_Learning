```python
from google.cloud import bigquery
```

```python
client=bigquery.Client()
```

<h1>Dataset</h1>
<h3>dataset= "nhtsa_traffic_fatalities"<br>
project= "bigquery-public-data"</h3>

```python
dataset_ref=client.dataset("nhtsa_traffic_fatalities",project="bigquery-public-data")
dataset=client.get_dataset(dataset_ref)
```

<h1>Table<\h1>
    <h3>table="accident_2015"</h3>

```python
table_ref=dataset_ref.table("accident_2015")
table=client.get_table(table_ref)
```

```python
preview_table=client.list_rows(table,max_results=5).to_dataframe()
```

```python
preview_table.head()
```

<h1>Query</h1>

```python
query='''
            Select COUNT(consecutive_number) AS num_accidents, EXTRACT(DAYOFWEEK FROM timestamp_of_crash) AS day_of_week
            FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
            GROUP BY day_of_week
            ORDER BY num_accidents DESC
            '''
```

```python
save_config=bigquery.QueryJobConfig(maximum_bytes=10**9)
query_job=client.query(query,job_config=save_config)
result=query_job.to_dataframe()
```
